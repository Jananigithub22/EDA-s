#### Let us check the maximum causes of death 

```python
plt.figure(figsize=(15,15))
plt.title('Types of causes for death', fontsize=20, fontweight='bold')
sns.countplot(y='Type', data=df, order=df['Type'].value_counts().index)
plt.xlabel('Count', fontsize=15, fontweight='bold')
plt.ylabel('Causes', fontsize=20, fontweight='bold')
plt.show()
```

* As observed above we can say that the maximum death is caused by 'OTHERS' 
* The minimum death is caused by three means of disase

#### Let us know which state has the maximum death rates

```python
df['State'].value_counts().plot(kind='bar',figsize=(15,7),colormap='ocean',fontsize=10)
plt.xlabel('STATES')
plt.ylabel('COUNT')
plt.title('BAR CAHRT SHOWING THE DEATH RATES IN DIFFERENT STATES')
plt.show()
```

```python
df['State'].value_counts()
```

As observed above we can say that the maximum death is in 3 States which is 
* Maharashtra           (6792)
* Madhya Pradesh        (6792)
* Karnataka             (6792)

#### Let us see the highest death rates among Male and Female

```python
p = sns.countplot(x='Gender', data = df, hue='Gender', palette='bright')
_ = plt.setp(p.get_xticklabels(), rotation=90) 
```
