Q1:what is percentage of crop yield of different states?

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x="State")
plt.title("crop Yield of different states")
plt.xticks(rotation=90)
```

Q2:which variety is most cultivating?

```python
plt.figure(figsize=(8,4))
sns.barplot(data=df,x=df["Variety"].head(20),y='Max Price')
plt.xticks(rotation=80)
```

Q3:What is count of the Grade?

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='Grade')
```

Q4:What is the maximum price in each state?

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x="State",y="Max Price")
plt.xticks(rotation=60)
plt.title("Max price in each state")
```

Q5:What is the modal price in each state?

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x="State",y="Modal Price")
plt.xticks(rotation=60) 
plt.title("Max price in each state")
```

Q6:What are the top 30 commodities?

```python
sns.histplot(data=df,x=df['Commodity'].head(30),kde=True)
plt.xticks(rotation=80)
```

Q7: Variations of Prize distribution?

```python
x=df[['Min Price',"Max Price","Modal Price"]]
plt.figure(figsize=(6,6))
sns.boxplot(data=x)
plt.title("Prize distribution")
```

Q8:Show the max price in each state in scatter plot?

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x='State',y="Max Price")
plt.xticks(rotation=80)
```

Q9:What are the commodities in top countries?

```python
top_data=df.nlargest(10,"Max Price")
plt.figure(figsize=(12,6))
sns.barplot(data=top_data,x="Max Price",y="State",hue="Commodity")
```

Q10:what is the commodities in the least countries ?

```python
small_data=df.nsmallest(10,'Modal Price')
plt.figure(figsize=(12,8))
sns.barplot(data=small_data,x='Max Price',y='State',hue='Commodity')
```
